```python
#chi_square_test
from scipy.stats import chi2_contingency

group1= data['region']
group2= data['Happiness Rank']

chi2, p, dof, expected = chi2_contingency(group1,group2)

print("Chi-Square Test:")
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)
alpha = 0.05  # Significance level
if p < alpha:
    print("Reject null hypothesis: There is a significant association between the variables.")
else:
    print("Fail to reject null hypothesis: No significant association between the variables.")
```
